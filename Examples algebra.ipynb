EXAMPLE VECTOR (ARRAY)

```python
v=np.array([1,2,3])
u=np.array([4,5,6])
```

```python
a=np.dot(v,u) #producte escalar
print(a)
```

```python
print(u)
```

```python
b=np.cross(v,u)  #producte vectorialS
print(b)
```

EXEMPLES MATRIUS

```python
c=np.array([u,v])
print(c)
```

```python
d=c[0] #coge la primera columna
print(d)
```

```python
e=c[0,0] #primero la fila y luego la columna
print(e)
```

```python
f=c[:,1]
print(f)
```

```python
A=np.array([u,v,u])
print(A)
```

VECTORS PROPIS

```python
valors,vectors=np.linalg.eig(A)  #enteindo q el orden es importante
print("valors propis", valors)
print("valors vectors", vectors)
```
